instance Functor ((->) r) where
    fmap :: (a -> b) -> (r -> a) -> r -> b 
    fmap f g = \x -> f (g x)
    --   ==    fmap = (.)
    --  first the value of type r (the third argument) needs to be transformed into a value of type a (by the r -> a function), so that the a -> b function can transform it into a value of type b (the result)

     return x = \_ -> x
     h >>= f = \w -> f (h w) w

# functor is “applicative” -> the contained value can be applied. In other words, it’s a FUNCTION.

offers a way to transform a contained function into one that can operate on contained values.

`(<*>) :: Applicative f -- for any applicative functor,
      => f (a ->   b)  -- take a contained function,
      -> (f a -> f b)  -- and make one that works on contained values`
      
Applicative functors also have a way to take an un-contained function and put it into a container:

`pure :: Applicative f -- for any applicative functor,
     =>   (a -> b)    -- take a normal function,
     -> f (a -> b)    -- and put it in a container`

In actuality, the type signature is simpler: `a -> f a`. Since `a` literally means “any type”, it can certainly represent the type `(a -> b)` too.

`pure :: Applicative f => a -> f a`

Even though the type signature for `(<*>)` starts with `f (a -> b)`, it can also be used with functions taking any number of arguments.

`:: f (a -> b -> c) -> f a -> f (b -> c)`

`(<*>) :: f (a -> ?) -> f a -> f ?`
substitute the type `(b -> c)` for every `?`




Curring 

When we say “a function of n arguments”, we’re actually lying. **_All functions in Haskell take exactly one argument_**.

Multi-argument functions are really single-argument functions that return other single-argument functions that accept the remaining arguments via the same process.

it’s entirely legal for that ? to be replaced with (b -> ?), and for that ? to be replaced with (c -> ?) and so on ad infinitum.

# applicative functors vs monads: Monad is for SERIES where Applicative is for PARALLEL.

This has nothing to do with concurrency or evaluation order

## Series vs Parallel

Using Monad means we have the ability to access the values returned by earlier lookup expressions in later ones.

```haskell
import Data.Text
import Data.String

--instance FromStr Text where
--  f x = x
-- deriving (IsString)
  
data User = User
    { userFirstName :: Text
    , userLastName  :: Text
    , userEmail     :: Text
    } deriving Show
-- f :: a    -> b    -> c    -> d
--User :: Text -> Text -> Text -> User

type Profile = [(Text, Text)]
-- Example:
-- [ ("first_name", "Pat"            )
-- , ("last_name" , "Brisbin"        )
-- , ("email"     , "me@pbrisbin.com")
-- ]
p :: Profile
p = []

lookup :: String -> Profile -> Maybe Text
lookup "last_name" _ = Just $ fromString "Dzbek"
lookup _ _ = Just $ fromString "Nothing"
-- x                  :: f     a
lookup "first_name" p :: Maybe Text

-- y                 :: f     b
lookup "last_name" p :: Maybe Text

-- z             :: f     c
lookup "email" p :: Maybe Text

-- result :: f d
-- result = f <$> x <*> y <*> z
buildUser :: Profile -> Maybe User
buildUser p = User
    <$> lookup "first_name" p
    <*> lookup "last_name" p
    <*> lookup "email" p
print "---"    
r = buildUser p   
:t r
r
-- r { userEmail = "Kupa"}
```
